```python
# introducing Matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Basic Graph
set1=[0,2,4,6] # X variables
set2= [0,8,10,12] # Y Variables
#       (X var, y var, label,   slope color,  Slope size, data points, Data point size, Color of data point) 
plt.plot(set1,set2, label="3x", color='blue', linewidth= 2, marker="*", markersize=10, markeredgecolor="red")

plt.title("Graph 1", fontdict={"fontname": "Calibri", "fontsize":20}) # font change and size
plt.xlabel("X side") 
plt.ylabel("Y side")
plt.xticks([0,1,2,3,4,5,6]) # x graph partition
plt.yticks([0,4,8,12]) # y graph partition
plt.legend()

plt.show() # display graph
```

```python
# Shorthand notation
set3=[0,1,2,3,4]
set4= [0,2,4,6,8]
plt.plot(set3,set4,"r^--", label="2x") # color,data point, slope line
plt.title("Graph 1", fontdict={"fontname": "Calibri", "fontsize":20})
plt.xlabel("X side")
plt.ylabel("Y side")
plt.xticks([0,1,2,3])
plt.yticks([0,4,8,12])
plt.legend()

plt.show()
```

```python
labels = ['Toyota', 'Honda', 'Nissan']
values = [2,4,1]

plt.figure(figsize=(5,3), dpi=100) # resolution and figure size

bars = plt.bar(labels, values)

patterns = ['-', '*', '/']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.savefig('barchart.png', dpi=300)

plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)
```

## Using Histogram

```python
bins = [40,50,60,70,80,90,100] # Partiton for ratings

plt.figure(figsize=(10,6))

plt.hist(fifa.Overall, bins=bins, color='blue')

plt.xticks(bins)

plt.ylabel('Player Numbers')
plt.xlabel('Ratings')
plt.title(' FIFA 2018 Player Skills distribution')

plt.savefig('histogram.png', dpi=300)

plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(10,6))

labels = ['Left', 'Right']
colors = ['blue', 'red']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%') # autocpt displays percentage in sectors

plt.title('Foot Preference of FIFA Players', fontdict={"fontweight": "bold", "fontsize": 20})


plt.show()
```

```python
plt.figure(figsize=(10,6), dpi=100)

plt.style.use('ggplot') # basic color for data visualization

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
# Conditional classification
light = fifa.loc[fifa.Weight < 125].count()[0] # Light weeight
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0] # light Medium
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0] # Medium weight
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0] # Medium Heavy
heavy = fifa[fifa.Weight >= 200].count()[0] # Heavy
# catrgories
weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,.1,.1,.4) # Spacing out sectors via label

plt.title('Weight of Professional Soccer Players in pounds')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```
